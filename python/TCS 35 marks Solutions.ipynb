```python
class Account:
    def __init__(self, CN, P, B, W, AT):
        self.CN = CN
        self.P = P
        self.B = B
        self.W = W
        self.AT = AT

    def update_balance(self, W1):
        if W1 <= self.B:
            self.B = self.B - W1
            self.W = W1


class ATM(Account):
    def __init__(self, l):
        self.l = l

    def ATM_Bal(self, CN1, P1, W2):
        for obj in self.l:
            if obj.CN == CN1 and obj.P == P1:
                obj.update_balance(W2)
                return obj
        return None

    def Filter_Ac(self, AT2):
        d = {}
        for obj in self.l:
            if obj.AT.lower() == AT2.lower():
                d[obj.CN] = obj.B
        dd = dict(sorted(d.items(), key=lambda x: x[1]))
        if dd == {}:
            return None
        else:
            return dd


l = []
n = int(input())
for _ in range(n):
    CN = int(input())
    P = int(input())
    B = float(input())
    W = float(input())
    AT = input()
    l.append(Account(CN, P, B, W, AT))

CN1 = int(input())
P1 = int(input())
W1 = float(input())
AT2 = input()

ATM_obj = ATM(l)
obj1 = ATM_obj.ATM_Bal(CN1, P1, W1)
if obj1 is not None:
    print(obj1.CN, obj1.B, obj1.W)
else:
    print("No account exists")

filter_acc = ATM_obj.Filter_Ac(AT2)
if filter_acc:
    for k, v in sorted(filter_acc.items()):
        print(k, v)
else:
    print("No match found")
```

```python
class RunningRace(Runner):
    def __init__(self,l):
        self.l=l
        
    def runner_pace(self,pace):
        l1=[]
        for i in self.l:
            pace1=i.T/i.Dis
            if pace1<pace:
                l1.append(i.NR)
        
        if l1==[]:
            return None
        else:
            return l1
    
    
```

```python
class Student:
    def __init__(self,R,N,l):
        self.R=R
        self.N=N
        self.l=l
        
    def cal_per(self):
        per = sum(self.l)/len(self.l)
        return per
    
    def find_grade(self,per):
        if per>=80:
            return "A"
        elif 80>per>=60:
            return "B"
        elif 60>per>=40:
            return "C"
        else:
            return "F"
        
R=int(input())
N=input()
n1=int(input())
l=[]
for i in range(n1):
    m=int(input())
    l.append(m)
    
o1=Student(R,N,l)
o2=o1.cal_per()
print(o2)
o3=o1.find_grade(o2)
print(o3)
```
