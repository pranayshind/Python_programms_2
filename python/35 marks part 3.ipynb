```python
class Employee:
    def __init__(self,N,D,S,OTC,OTS=False):
        self.N=N
        self.D=D
        self.S=S
        self.OTC=OTC
        self.OTS=OTS
    
class Organization:
    def __init__(self,l):
        self.l=l
        
    def Eligible(self,thre):
        for i in self.l:
            if sum(i.OTC.values())>=thre:
                i.OTS=True
            else:
                i.OTS=False
            print(i.N,i.OTS)
    
    def Bonus(self,rate):
        l1=[]
        for i in self.l:
            if i.OTS==True:
                l1.append(sum(i.OTC.values())*rate)
        return sum(l1)
                
l=[]
n=int(input())
for i in range(n):
    N=input()
    D=input()
    S=input()
    n2=int(input())
    OTC={}
    for j in range(n2):
        month=input()
        ot=int(input())
        OTC[month]=ot
    l.append(Employee(N,D,S,OTC))
    
thre=int(input())
rate=int(input())

o1=Organization(l)
o2=o1.Eligible(thre)
print(o2)
o3=o1.Bonus(rate)
print(o3)
```
