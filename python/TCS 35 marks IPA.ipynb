```python
class Account:
    def __init__(self, cardNumber, pin, balance, withdrawalAmount, accountType):
        self.cardNumber = cardNumber
        self.pin = pin
        self.balance = balance
        self.withdrawalAmount = withdrawalAmount
        self.accountType = accountType

    def update_balance(self, withdrawalAmount):
        if withdrawalAmount <= self.balance:
            self.balance -= withdrawalAmount
            self.withdrawalAmount = withdrawalAmount

class ATM:
    def __init__(self, l):
        self.l = l

    def ATM_Balance(self, card_number, pin, withdrawalAmount):
        for obj in self.l:
            if obj.cardNumber == card_number and obj.pin == pin:
                obj.update_balance(withdrawalAmount)
                return obj
        return None

    def ATM_Dic(self, accountType):
        d = {}
        for obj in self.l:
            if obj.accountType.lower() == accountType.lower():
                d[obj.cardNumber] = obj.balance
        return d

l = []  # list of accountObject
n = int(input())
for _ in range(n):
    cardNumber = int(input())
    pin = int(input())
    balance = float(input())
    withdrawalAmount = float(input())
    accountType = input()
    l.append(Account(cardNumber, pin, balance, withdrawalAmount, accountType))

card_number = int(input())
pin = int(input())
withdrawalAmount = float(input())
accountType = input()

ATM_obj = ATM(l)
account = ATM_obj.ATM_Balance(card_number, pin, withdrawalAmount)
if account is not None:
    print(account.cardNumber, account.balance, account.withdrawalAmount)

filter_acc = ATM_obj.ATM_Dic(accountType)
if filter_acc:
    for cardNumber, balance in sorted(filter_acc.items(), key=lambda x: x[1]):
        print(cardNumber, balance)
else:
    print("No Match Found")
```
